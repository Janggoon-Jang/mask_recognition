1. 라벨링 품질 확인
2. 이미지 형식 통일(jpg), 손상 파일 제거
3. 파일명 변환

```python
import cv2
import matplotlib.pyplot as plt
import os
import shutil
```

# 1. 라벨링 품질 확인

```python
root_dir = '/mask_recognition/data/dataset_kaggle_Face_Mask_Dataset_YOLO_Format/data'

i = 0
for filepath in os.listdir(root_dir):
    if i > 50:
        break
    if not filepath.endswith('.jpg'):
        continue
    
    filename = os.path.splitext(filepath)[0]
    imagepath = os.path.join(root_dir, f"{filename}.jpg")
    labelpath = os.path.join(root_dir, f"{filename}.txt")

    img = cv2.imread(imagepath)
    h, w, _ = img.shape

    with open(labelpath) as f:
        lines = f.readlines()
        for line in lines:
            label, cx, cy, bx, by = [float(value) for value in line.split()]

            x1 = int((cx - bx/2) * w)
            x2 = int((cx + bx/2) * w)
            y1 = int((cy - by/2) * h)
            y2 = int((cy + by/2) * h)

            cv2.rectangle(img, (x1, y1), (x2, y2), (255, 0, 0), 3)

    plt.imshow(img)
    plt.show()
    i += 1 
```

# 2. 이미지 형식 통일(.jpg),손상 파일 제거

1. 이미지 파일이 잘 열리는지 검사
2. 열리지 않는 파일 삭제
3. jpg가 아닌 이미지 파일 jpg로 변환
4. 변환된 이미지의 원본 파일 삭제

```python
data_dir = '/mask_recognition/data/dataset_kaggle_Face_Mask_Dataset_YOLO_Format/data'

unreadible = []
not_jpg = []

for file in os.listdir(data_dir):
    filepath = os.path.join(data_dir, file)
    name, ext = os.path.splitext(file)

    if ext == '.txt':
        continue

    image = cv2.imread(filepath)

    # 읽지 못 하는 이미지 파일 이름 저장
    if image is None:
        unreadible.append(file)
        print(f'{file} is not readible.')
        continue

    # jpg가 아닌 정상 이미지 변환
    if ext != '.jpg':
        not_jpg.append(file)
        new_path = os.path.join(data_dir, f'{name}.jpg')
        new_file = cv2.imwrite(new_path, image)
        if new_file:
            print(f'{file} is changed to jpg.')

            # jpg로 정상 변환된 원본 이미지 삭제
            try:
                os.remove(filepath)
                print(f'{filepath} deleted.')
            except Exception as e:
                print(f'Failed to delete {file} : {e}')
        else:
            print(f"failed to write {file} as jpg.")

print(f'{len(unreadible)} images are unreadible.')
print(f'{len(not_jpg)} images are not jpg.')
```

# 3. 파일명 변환

1. 이미지 파일의 이름을 {i}.jpg로 저장
2. 같은 이름의 txt 파일의 이름을 {i}.txt로 저장

```python
src_dir = '/mask_recognition/data/dataset_kaggle_Face_Mask_Dataset_YOLO_Format/data'
dst_dir = '/mask_recognition/data/dataset_kaggle_Face_Mask_Dataset_YOLO_Format/renamed'
os.makedirs(dst_dir, exist_ok=True)

jpg_list = sorted([file for file in os.listdir(src_dir) if file.endswith('.jpg')])

success = 0
failed = 0
for i, jpg in enumerate(jpg_list):
    name, ext = os.path.splitext(jpg)

    jpg_path = os.path.join(src_dir, jpg)
    txt_path = os.path.join(src_dir, f'{name}.txt')
    
    if not os.path.exists(txt_path):
        print(f'Not exist : {txt_path}')
        failed += 1
        continue

    new_jpg = os.path.join(dst_dir, f'{i}.jpg')
    new_txt = os.path.join(dst_dir, f'{i}.txt')

    try:
        shutil.move(jpg_path, new_jpg)
        shutil.move(txt_path, new_txt)
        success += 1
    except Exception as e:
        print(f'{e} : Failed to move {jpg_path} or {txt_path}')
        failed += 1

print(f'Converted files : {success}')
print(f'Failed files : {failed}')
```
